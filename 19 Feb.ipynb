#Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a web framework for Python, designed to make it easy to build web applications. It is classified as a microframework because it does not require or impose any particular tools or libraries. Instead, it focuses on simplicity and minimalism while providing the necessary features for web development.

Here are some advantages of using the Flask framework:

Lightweight and Simple: Flask is designed to be lightweight and has a small core, which makes it easy to learn and use. It doesn't come bundled with unnecessary features or dependencies, allowing developers to have more control over their application structure.

Flexibility: Flask gives developers the freedom to choose the tools and libraries they want to work with. It does not enforce any specific way of doing things, allowing developers to make decisions based on their preferences and project requirements.

Easy to Get Started: Flask has a straightforward and intuitive API that makes it easy to get started with web development. It has a minimalistic design and doesn't require much boilerplate code, which reduces the learning curve for beginners.

Extensibility: Flask is highly extensible and provides a wide range of extensions that can be easily integrated into applications. These extensions cover various functionalities such as database integration, form validation, authentication, and more, allowing developers to add extra features without reinventing the wheel.

Modular and Scalable: Flask follows a modular design, allowing developers to build applications in a structured manner. It supports modular application structures and follows the WSGI (Web Server Gateway Interface) standard, which enables easy integration with other web servers or applications.

#Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or view handlers in your application. It allows you to define different routes for different URLs and define the corresponding actions or responses to be taken when those URLs are accessed.

The Flask app.route() decorator is used to define routes in Flask applications. It is a way to bind a URL pattern to a view function, which is responsible for handling the request and generating the response.

#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.


In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint (function/view) in your Flask application. This function helps you avoid hard-coding URLs in your templates or application code, making it easier to maintain and update URLs.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    # Generate a URL for the 'about' endpoint
    about_url = url_for('about')
    return f'The URL for the About page: {about_url}'

@app.route('/about')
def about():
    return 'This is the About page.'

if __name__ == '__main__':
    app.run()
